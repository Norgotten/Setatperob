### Mencari range, IQR, varian, standar deviasi

```python
range1 = df["Penjualan A (pcs)"].max() - df["Penjualan A (pcs)"].min()
range2 = df["Penjualan B (pcs)"].max() - df["Penjualan B (pcs)"].min()
range3 = df["Curah Hujan (mm)"].max() - df["Curah Hujan (mm)"].min()

Q1_a = df["Penjualan A (pcs)"].quantile(0.25)
Q3_a = df["Penjualan A (pcs)"].quantile(0.75)
IQR_a = Q3_a - Q1_a

Q1_b = df["Penjualan B (pcs)"].quantile(0.25)
Q3_b = df["Penjualan B (pcs)"].quantile(0.75)
IQR_b = Q3_b - Q1_b

Q1_3 = df["Curah Hujan (mm)"].quantile(0.25)
Q3_3 = df["Curah Hujan (mm)"].quantile(0.75)
IQR_3 = Q3_3 - Q1_3

varian1 = df["Penjualan A (pcs)"].var()
varian2 = df["Penjualan B (pcs)"].var()
varian3 = df["Curah Hujan (mm)"].var()

std_dev1 = df["Penjualan A (pcs)"].std()
std_dev2 = df["Penjualan B (pcs)"].std()
std_dev3 = df["Curah Hujan (mm)"].std()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Path file Excel
file_path = r'C:\Users\Mass Zee\statprob\Dataset.xlsx'

# Membaca file Excel
data = pd.read_excel(file_path)

# Function to create frequency distribution, histogram, and frequency polygon
def plot_distribution_and_histogram(column, col_name):
    # Creating frequency distribution (bins and frequency count)
    freq_dist = pd.cut(column, bins=10).value_counts().sort_index()

    # Plotting histogram
    plt.figure(figsize=(10, 6))
    
    # Histogram
    plt.subplot(1, 2, 1)
    plt.hist(column, bins=10, edgecolor='black')
    plt.title(f'Histogram {col_name}')
    plt.xlabel(col_name)
    plt.ylabel('Frekuensi')

    # Frequency Polygon
    plt.subplot(1, 2, 2)
    bin_edges = np.histogram_bin_edges(column, bins=10)
    hist, _ = np.histogram(column, bins=bin_edges)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.plot(bin_centers, hist, marker='o', linestyle='-', color='blue')
    plt.title(f'Poligon Frekuensi {col_name}')
    plt.xlabel(col_name)
    plt.ylabel('Frekuensi')

    plt.tight_layout()
    plt.show()
    
    return freq_dist

# For Curah Hujan
curah_hujan_freq = plot_distribution_and_histogram(data['Curah Hujan (mm)'], 'Curah Hujan (mm)')

# For Penjualan A
penjualan_A_freq = plot_distribution_and_histogram(data['Penjualan A (pcs)'], 'Penjualan A (pcs)')

# For Penjualan B
penjualan_B_freq = plot_distribution_and_histogram(data['Penjualan B (pcs)'], 'Penjualan B (pcs)')

# Display the frequency distribution tables
print(curah_hujan_freq)
print(penjualan_A_freq)
print(penjualan_B_freq)
```

Descriptive Analysis:
Curah Hujan (mm) 
   Rata-rata curah hujan adalah 14.26 mm, dengan standar deviasi 27.54 mm, menunjukkan adanya variabilitas yang tinggi.
   Nilai minimum curah hujan adalah 0 mm, dan nilai maksimumnya mencapai 170.6 mm, dengan 25% data memiliki nilai 0 mm (cukup banyak hari tanpa hujan).
  
Penjualan A (pcs)

Penjualan B (pcs)

Outliers:
Curah Hujan (mm) memiliki 13 outliers. Ini menunjukkan bahwa terdapat hari-hari dengan curah hujan yang sangat ekstrem (jauh di atas rata-rata), kemungkinan hari dengan hujan sangat deras.
Penjualan A dan Penjualan B tidak menunjukkan adanya outlier yang signifikan menurut metode IQR, sehingga persebaran data penjualan ini lebih stabil tanpa nilai yang sangat ekstrem.
